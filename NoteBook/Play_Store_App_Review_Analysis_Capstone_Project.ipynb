## Taking Care of Null values

```python
# checking the number of null values presented in each column

df.isna().sum()
```

To get the max results from our analysis, we are filling null values of 'Rating' with median Ratings of its respective 'Category'.

```python
# to fill with 'Category' median, we are using groupby and transform method

df["Rating"] = df.groupby("Category")['Rating'].transform(lambda x: x.fillna(x.median()))
```

Now filling null values of 'Type',  'Current Ver',  'Android Ver' columns with their modes respectively.

```python
# filling mode of themselves in null values of columns using fillna method

df['Type'].fillna(df['Type'].mode()[0], inplace=True) # we used [0] because we are taking value at 0th index of the mode() output
df['Current Ver'].fillna(df['Current Ver'].mode()[0], inplace=True)
df['Android Ver'].fillna(df['Android Ver'].mode()[0], inplace=True)
```

```python
df.info()
```

as seen in info, Object is the Data Type of following column- 

*   Price
*   Review
*   Installs

we need to change it to their preferred type of integer and float

```python
# for Price
# 1. removing the '$' sign in 'Price' column
df['Price'] = df['Price'].apply(lambda x: str(x).replace('$', '') if '$' in str(x) else str(x))
# 2. converting the string type price to float type 
df['Price'] = df['Price'].apply(lambda x: float(x))


# for Review, we use pandas 'to neumeric' method cause it has large numbers
df['Reviews'] = pd.to_numeric(df['Reviews'],errors='coerce')


# for Installs, 
# 1. we first replace unwanted characters from numeric string
df['Installs'] = df['Installs'].str.replace('+','').str.replace(',','').str.replace('-','')
# 2. now we convert it to integer using apply method
df['Installs'] = df['Installs'].apply(lambda x: int(x))
```

## Values of 'Size' column in data are of string format and in MB or KB so we are converting them to MB for optimum results.

```python
df['Size']= df['Size'].apply(lambda x: str(x).replace('M', '') if 'M' in str(x) else x) #     Replacing 'M' from string of Size
df['Size']= df['Size'].apply(lambda x: str(x).replace('Varies with device', 'NaN') if 'Varies with device' in str(x) else x)
df['Size']= df['Size'].apply(lambda x: float(str(x).replace('k', ''))/1000 if 'k' in str(x) else x) #converting Kb of size to MB by dividing to 1000
df['Size']= df['Size'].apply(lambda x: float(x)) # converting all values to float

df['Size'].fillna(float(int(df['Size'].mean())), inplace=True) # filled nan values of size with the mean of size
```

## Analysing the No. of Positive, Negative and Neutral Sentiments in each category

```python
# creating temporaary df to store only required columns
tdf = new_df.loc[:,['Category','Sentiment']]

# now structuring our required data, grouping and counting values of each sentimewnt for each catalogue so that we can plot on graph
tdf = tdf.groupby('Category')['Sentiment'].value_counts().unstack()

# plotting the groupped bar graph
plt.rcParams["figure.figsize"] = (15,15)
tdf.plot(kind='barh',width = 0.8, stacked=True)
```

Insights:-

>In reviews, overall positive sentiments are more compare to negative and neutral are less, and specifically Gaming has largest reviews with positive and negative sentiment of approx 60% and 40% respectively, all other category has following the same pattern of positive reviews more than negetive reviews and neutral are the lowest, by which we conclude that none of category is performing negative particularly.



## Gathering the Percentage of Positive sentiments for every Category high to low

```python
# counting categoies of ony positive sentiments and assign to new dataframe
tdf1 = pd.DataFrame(new_df[new_df['Sentiment']=='Positive']['Category'].value_counts().sort_values(ascending=False)).reset_index()

# counting categoies of ony Negetive sentiments and assign to new dataframe
tdf2 = pd.DataFrame(new_df[new_df['Sentiment']=='Negative']['Category'].value_counts().sort_values(ascending=False)).reset_index()

# counting categoies of ony Negetive sentiments and assign to new dataframe
tdf3 = pd.DataFrame(new_df[new_df['Sentiment']=='Neutral']['Category'].value_counts().sort_values(ascending=False)).reset_index()

tdf = tdf1.merge(tdf2, on='index') #merging the 3 data frames
tdf = tdf.merge(tdf3, on='index')

tdf.columns=['Category','Positive','Negative','Neutral'] #changing column names

# adding new column in temp data frame of percentage where we store percentage of positive sentiments of each category
tdf['Positive percentage'] = round((tdf['Positive']/(tdf['Positive']+tdf['Negative']+tdf['Neutral']))*100,2)

tdf = tdf.sort_values(['Positive percentage'],ascending=False) #sorted the percentage descending
```

```python
# plotting the temp data frame
plt.plot(tdf['Category'], tdf['Positive percentage'],'r-o')
plt.style.use('fivethirtyeight')

plt.tick_params(axis='x', colors='red')
plt.tick_params(axis='y', colors='red')

plt.yticks(fontsize= 18)
plt.xticks(fontsize= 18, rotation=90)

plt.title('Percentage of Positive sentiments for each Category ',color='black')
plt.xlabel('Category >>>>',color='black')
plt.ylabel('Positive Percentage >>>', color='black')

plt.show()
```

Insights:-

> From the graphical representation, there are 5 categories which has more than 75% of positive reviews of total reviews and we can conclude that these categories are already satisfying its users, and the categories with less positivity are yet to receive the user satisfaction, because a negative review are refrence to dissatisfied user.
